```python
#Problem1:1D Random Walk
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
Numsteps = 100     # no. of steps
Numexp=1000      # total number of experiments
Avgdis=[]   
for steps in range(Numsteps):
    SED=[]      
    for i in range(Numexp):
        x=0 # start from origin
        for j in range(steps):
            x += random.choice([1, -1])  # one step jump either left(-1) or right(+1)
        SED.append(np.abs(x))            # Absolute distances for any steps
    Avgdis.append((np.mean(SED)))        # mean distance for all steps
plt.title('1D Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Average Distance')
plt.plot(range(Numsteps), Avgdis,'r');
```

```python
#Problem2:1D Biased Random Walk
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
Numsteps = 100      # no. of steps
Numexp=1000       # total number of experiments
Avgdis=[]   
for steps in range(Numsteps):
    SED=[]      
    for i in range(Numexp):
        x=0 # start from origin
        for j in range(steps):
            x += random.choice([-1,1,-1, -1])  # three times likely to jump left than right
        SED.append(np.abs(x))                  # Absolute distances for any steps
    Avgdis.append((np.mean(SED)))              # mean distance for all steps
plt.title('1D Biased Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Average Distance')
plt.plot(range(Numsteps), Avgdis, 'b');
```

```python
#Problem3:3D Random Walk
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
Numsteps = 100      # no. of steps
Numexp=1000         # total number of experiments
Avgdis=[]   
for steps in range(Numsteps):
    SED=[]
    distance=0
    for i in range(Numexp):
        x,y,z=(0,0,0) # start from origin
        for j in range(steps):
            jump3D = random.choice([1,2,3,4,5,6])# jump in any three direction from a point
            if jump3D==1:
                x=x+1
            elif jump3D==2:
                x=x-1
            elif jump3D==3:
                y=y+1
            elif jump3D==4:
                y=y-1
            elif jump3D==5:
                z=z+1
            else:
                z=z-1
            distance=np.sqrt(x*x+y*y+z*z)      # distance from origion
        SED.append(distance)                
    Avgdis.append((np.mean(SED)))              # mean distance for any steps
plt.title('3D Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Average Distance')
plt.plot(range(Numsteps), Avgdis, 'g');
```

```python
#Problem4:3D Random Walk with Random Direction
import math
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
Numsteps =10    # no. of steps
Numexp=1000      # total number of experiments
Avgdis=[]   
for steps in range(Numsteps):
    SED=[]
    distance=0
    for i in range(Numexp):
        # start from origin
        x=0
        y=0
        z=0
        thetastep=(180/Numexp)*math.pi/180
        phistep=(360/Numexp)*math.pi/180
        theta=0
        phi=0
        for j in range(steps):
            jump3D = random.choice([1,2,3,4])# jump in any three direction from a point
            if jump3D==1:
                theta=theta + thetastep
            elif jump3D==2:
                theta=theta - thetastep
            elif jump3D==3:
                phi=phi + phistep
            else:
                phi=phi - phistep

                theta=theta*180/math.pi
                phi=phi*180/math.pi
                x=np.sin(theta)* np.cos(phi)
                y=np.sin(theta)* np.sin(phi)
                z=np.cos(theta)
            distance=np.sqrt(x*x+y*y+z*z)      # distance from origion
        SED.append(distance)                
    Avgdis.append((np.mean(SED)))              # mean distance for any steps
plt.title('3D Random Walk with Random Direction')
plt.xlabel('Number of Steps')
plt.ylabel('Average Distance')
plt.plot(range(Numsteps), Avgdis, 'y');
```
