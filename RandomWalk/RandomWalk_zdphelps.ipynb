```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random
%matplotlib inline

#Initial Conditions
N = 1000
val_array = np.array([0])
AbsVal_array = np.array([0])
Average_array = np.array([25.0])
#Loop for 50 averages
for i in range(50):
    Total = 0
    #Loop for increaseing number of particles
    for j in range(10*(i+1)):
        move = 0
        val = 0
        #loop for movement on a 1D path
        for k in range(N):    
            #fair choice
            move = random.randint(0,2)
            #biased choice
            #move = random.choice([0,1,1,1])
            
            #updates position based on move
            if(move == 0):
                val = val - 1
            if(move == 1):
                val = val + 1
        #appends final and absloute value position to array
        val_array = np.append(val_array,val)
        AbsVal_array = np.append(AbsVal_array,abs(val))
        #updates total (to average)
        Total = Total + AbsVal_array[j]
    #calculates average for interval and appends to array
    Average = Total/(10.0*(i+1))
    Average_array = np.append(Average_array,Average)



#plots figures   
plt.figure()
plt.plot(val_array,'.')

plt.figure()
plt.hist(val_array,100)

plt.figure()
plt.plot(AbsVal_array,'.')

plt.figure()
plt.hist(AbsVal_array,100)

plt.figure()
plt.plot(Average_array,'-')

#the average array looks strange because the inital value of the average array is initialized to an arbitrary
#value. I put in a value that is within the general average for the non biased case; however, the biased case
#looks drastically different because it is counting that inital value (I wasnt sure how to inialize an array
# without a given value)
```
