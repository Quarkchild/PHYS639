```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random
%matplotlib inline

N = 1000
val_array = np.array([0])
AbsVal_array = np.array([0])
Average_array = np.array([25.0])

for i in range(50):
    Total = 0
    for j in range(10*(i+1)):
        move = 0
        val = 0
        for k in range(N):    
            move = random.randint(0,2)
            if(move == 0):
                val = val - 1
            if(move == 1):
                val = val + 1
        #print(val)
        val_array = np.append(val_array,val)
        AbsVal_array = np.append(AbsVal_array,abs(val))
        Total = Total + AbsVal_array[j]
    Average = Total/(10.0*(i+1))
    Average_array = np.append(Average_array,Average)



#x = N/10    
plt.figure()
plt.plot(val_array,'.')

plt.figure()
plt.hist(val_array,100)

plt.figure()
plt.plot(AbsVal_array,'.')

plt.figure()
plt.hist(AbsVal_array,100)

plt.figure()
plt.plot(Average_array,'-')
```
