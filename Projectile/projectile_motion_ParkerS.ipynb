```python
#Written by Parker Stoops
#
#
#Compute how far a projectile flies before hitting the ground, ignoring all other effects besides gravity
#Drag Force->      F/m=-.00004v^2
#

import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline

x_int = 0 #starting distance (m)
y_int = 0 #starting height (m)
t_int = 0 #starting time (sec)
g = -9.8  #gravity (m/s^2)
m = 50   #mass of projectile
v_int = 100 #intial velocity (m/s)
S = 1000000 #Stepsize
t_int = 0
t_fin = 1000 #time in (sec)
dt = (t_int + t_fin) /(S)
deg = 45 #intial lauch angle
Theta = deg * np.pi / 180 #In radians
v_xint = v_int * np.cos(Theta)
v_yint = v_int * np.sin(Theta)

x_array = np.array([x_int])
y_array = np.array([y_int])

while y_int >= 0:
    
    v_yint = v_yint + g*dt
    v_xint = v_xint #gravity has no effect
    x_int = x_int + v_xint*dt
    y_int = y_int + v_yint*dt
    t_int = t_int + dt
    x_array = np.append(x_array,x_int)
    y_array = np.append(y_array,y_int)

plt.plot(x_array,y_array,label='proj')
plt.xlabel('x (m)')
plt.ylabel('y (m)')



print('The distance that it went is', np.max(x_array), 'meters')
print('The height that it reached was', np.max(y_array), 'meters')





```

```python
# #2  With Drag
# 
#Drag Force->      F/m=-.00004v^2
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline

x_int = 0 #starting distance (m)
y_int = 0 #starting height (m)
t_int = 0 #starting time (sec)
g = -9.8  #gravity (m/s^2)
mass = 50   #mass of projectile
v_int = 100 #intial velocity (m/s)
S = 1000000 #Stepsize
t_int = 0
t_fin = 1000 #time in (sec)
dt = (t_int + t_fin) /(S)
deg = 45 #intial lauch angle
theta = deg * np.pi / 180 #In radians
v_xint = v_int * np.cos(Theta)
v_yint = v_int * np.sin(Theta)
F_xdg = (-mass * 0.00004 * v_xint**2) * np.cos(theta)
F_ydg = (-mass * 0.00004 * v_yint**2) * np.sin(theta)

x_array = np.array([x_int])
y_array = np.array([y_int])

while y_int >= 0:
    a_x = (F_xdg/mass)
    a_y = g + (F_ydg / mass)
    v_yint = v_yint + a_y*dt
    v_xint = v_xint  + a_x*dt #gravity has no effect
    x_int = x_int + v_xint*dt
    y_int = y_int + v_yint*dt
    t_int = t_int + dt
    x_array = np.append(x_array,x_int)
    y_array = np.append(y_array,y_int)

plt.plot(x_array,y_array,label='proj_drag')
plt.xlabel('x (m)')
plt.ylabel('y (m)')



print('The distance that it went is', np.max(x_array), 'meters')
print('The height that it reached was', np.max(y_array), 'meters')

```
