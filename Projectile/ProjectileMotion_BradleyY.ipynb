```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Simple projectile motion level (*)
##############################################
#Initial Conditions
m=1.0
t=0.0 #Initial Time
x=0.0#Initial x-position (In meters)
y=0.0 #Initial y-position (In meters)
v=100.0#Initial Velocity (In meters)
theta=np.deg2rad(45)#Initial launch angle converted from degrees to radians
vx=v*np.cos(theta)
vy=v*np.sin(theta) #Defining the x and y components of v0
dt=0.001 # Time increments
Fxdrag=(-m*0.00004*vx**2)*np.cos(theta)
Fydrag=(-m*0.00004*vy**2)*np.sin(theta)

xarr=np.array([x])
yarr=np.array([y])#Creating an x and y array for futur element inputs (Credit due: Lado Samushia)
```

```python
#Numerically Solve the differential equations
while y>= 0: #A while loop that will compute the evolution of projectile, We want the projectile to stay at or above y=0
    ax=(0*(Fxdrag/m))
    ay=-9.8+(0*(Fydrag/m))
    x+=vx*dt
    y+=vy*dt
    vx+=ax*dt
    vy+=ay*dt
    t+=dt
    xarr=np.append(xarr,x)
    yarr=np.append(yarr,y)#Appending new value to the array (Credit due:Lado Samushia)
##############################################################################################
#Air Resistance Where Fdrag/m=-0.000004*v**2
```

```python
plt.plot(xarr,yarr,label='Trajectory')
plt.xlabel('X Range (m)')
plt.ylabel('Y Range (m)')
```

```python
m2=1.0
t2=0.0 #Initial Time
x2=0.0#Initial x-position (In meters)
y2=0.0 #Initial y-position (In meters)
v2=100.0#Initial Velocity (In meters)
theta2=np.deg2rad(45)#Initial launch angle converted from degrees to radians
vx2=v2*np.cos(theta)
vy2=v2*np.sin(theta) #Defining the x and y components of v0
dt2=0.001 # Time increments
Fxdrag2=(-m2*0.00004*vx2**2)*np.cos(theta2)
Fydrag2=(-m2*0.00004*vy2**2)*np.sin(theta2)

xarr2=np.array([x2])
yarr2=np.array([y2])#Creating an x and y array for futur element inputs (Credit due: Lado Samushia)
```

```python
while y2>= 0: #A while loop that will compute the evolution of projectile, We want the projectile to stay at or above y=0
    ax2=((Fxdrag2/m2))
    ay2=-9.8+((Fydrag2/m2))
    x2+=vx2*dt2
    y2+=vy2*dt2
    vx2+=ax2*dt2
    vy2+=ay2*dt2
    t2+=dt2
    xarr2=np.append(xarr2,x2)
    yarr2=np.append(yarr2,y2)#Appending new value to the array (Credit due:Lado Samushia)
```

```python
plt.plot(xarr2,yarr2,label='Air Resistance Trajectory')
plt.xlabel('X Range (m)')
plt.ylabel('Y Range (m)')
```
