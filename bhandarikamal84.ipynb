```python
# PROBLEM: RADIOACTIVE DECAY
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(tau): # Define function with half life as a variable
    N0=1000
    tini=0
    tfin=10*tau
    Nsteps=10000
    t=np.linspace(tini,tfin,Nsteps+1)
    N=np.zeros(Nsteps+1)
    N[0]=N0
    dt=(tfin-tini)/Nsteps
    for i in range(1,Nsteps):
        dN=-N[i-1]/tau
        N[i]=N[i-1]+dN*dt
    plt.plot(t,N)
    plt.xlabel('time')
    plt.ylabel('Numbers of atoms');

   
```

```python
f(500) # plot 1 blue
f(700)  # plot 2 yellow
print("This plot shows that number of atoms in radioactive material decays exponentially with time.")
print("Also shorter the half time, steeper the rate through which it decays.")
```

```python
#PROBLEM: POPULATION GROWTH
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(A,B,N0):  # define function with A,B and initial number of atoms as variables
    tini=0
    tfin=10
    Nsteps=10000
    t=np.linspace(tini,tfin,Nsteps+1)
    N=np.zeros(Nsteps+1)
    N[0]=N0
    dt=(tfin-tini)/Nsteps
    for i in range(1,Nsteps+1):
        dN= A*N[i-1]- B* N[i-1]*N[i-1]
        N[i]=N[i-1]+dN*dt
    plt.plot(t,N)
    plt.xlabel('time')
    plt.ylabel('Numbers of atoms');
```

```python
f(0.5,0.02,15)  #plot1 blue
f(15,1,15)      #plot2 yello
f(10,1,15)      #plot3 green
print("If A > B N, population increases with time.")
print("If A = B N, equilibrium condition is acheived.")
print("If A < B N, population decreases  with time.")
```

```python
# Problem: COUPLED RADIOACTIVE DECAY1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(tauA,tauB): # Define function with half life of both material
    N00=100
    N10=0
    tini0=0
    tini1=0
    tfin0=500
    tfin1=500
    Nsteps=10000
    t0=np.linspace(tini0,tfin0,Nsteps+1)
    t1=np.linspace(tini1,tfin1,Nsteps+1)
    N0=np.zeros(Nsteps+1)
    N1=np.zeros(Nsteps+1)
    N0[0]=N00
    N1[0]=N10
    dt0=(tfin0-tini0)/Nsteps
    dt1=(tfin1-tini1)/Nsteps
    for i in range(1,Nsteps+1):
        dN0=-N0[i-1]/tauA
        N0[i]=N0[i-1]+dN0*dt0
        dN1=N0[i-1]/tauA-N1[i-1]/tauB
        N1[i]=N1[i-1]+dN1*dt1
    plt.plot(t0,N0)
    plt.plot(t1,N1)
    plt.xlabel('time')
    plt.ylabel('Numbers of atoms');

```

```python
f(20,10) # partA
print("Atoms A always decays but Atoms B initially increses but starts to decay after sometimes.")
print("Finally they reach at equilibrium.")
print("But if we increase tauA, atoms A decays slower and vice-versa.")
print("and if we increase tauB, Atoms B initially increase more, then starts to decay and vice-versa.")
```

```python
# Problem: COUPLED RADIOACTIVE DECAY2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(N00,N10): # Define function with initial number of atoms of both material
    tau=10
    tini0=0
    tini1=0
    tfin0=10*tau
    tfin1=10*tau
    Nsteps=10000
    t0=np.linspace(tini0,tfin0,Nsteps+1)
    t1=np.linspace(tini1,tfin1,Nsteps+1)
    N0=np.zeros(Nsteps+1)
    N1=np.zeros(Nsteps+1)
    N0[0]=N00
    N1[0]=N10
    dt0=(tfin0-tini0)/Nsteps
    dt1=(tfin1-tini1)/Nsteps
    for i in range(1,Nsteps+1):
        dN0=N1[i-1]/tau-N0[i-1]/tau
        N0[i]=N0[i-1]+dN0*dt0
        dN1=N0[i-1]/tau-N1[i-1]/tau
        N1[i]=N1[i-1]+dN1*dt1
    plt.plot(t0,N0)
    plt.plot(t1,N1)
    plt.xlabel('time')
    plt.ylabel('Number of atoms');
```

```python
f(1000,500) # plot1 blue and yellow
f(100,200)  # plot2 green and red

print("If tau is same for both material, material with large initial number of atoms starts to decay and material with")
print("less initial atoms starts to increase but eventually they get euilibrium condition.")
```
