```python
#Problem1:Infinite Potential Wall
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# Assume
Nsteps=1000
dx=L/Nsteps
Eini= 4 
dE=0.01
Emax=150
# Given
L=1 # Unit length of well
hcut=1 # Planck's constant
m=1 # unit mass
V=0 # in region 0<X<L
#Define
pi=np.pi
PsiA=np.zeros(Nsteps)
dPsiA=np.zeros(Nsteps)
Psi_final=[]
Energy_final=[]
while Eini<Emax:
    #initial conditions
    Psi=0
    dPsi=1
    for j in range(Nsteps):
        ddPsi  = -2*Eini*Psi 
        dPsi +=  ddPsi*dx
        Psi  +=   dPsi*dx
        
        PsiA[j]= Psi
        dPsiA[j]= dPsi
    
    if PsiA[Nsteps-2]>0 and PsiA[Nsteps-1]<0 or PsiA[Nsteps-2]<0 and PsiA[Nsteps-1]>0:
        Psi_final.append(Psi)
        Energy_final.append(Eini)
        print(Eini)
        Eini = Eini + (0.01)*Eini
        
        plt.figure(1)
        plt.plot(PsiA)
        plt.xlabel('X')
        plt.ylabel('Psi')
        plt.title('Wavefunctions')
        
        plt.figure(2)
        plt.plot(PsiA**2)
        plt.xlabel('X')
        plt.ylabel('Psi Square')
        plt.title('Probability distribution function')
        
        plt.figure(3)
        plt.plot(Energy_final,'o')
        plt.xlabel('Energy level')
        plt.ylabel('Energy ')
        plt.title('Discrete Energy Level')
        
    else:
        Eini=Eini+dE   
       
```
