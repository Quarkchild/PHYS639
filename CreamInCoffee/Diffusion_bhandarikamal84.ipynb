```python
#Problem1: Cream in Coffee
import math
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
%matplotlib inline
Nparticles = 400
Ngrid= 100
X= np.zeros(Nparticles)
Y= np.zeros(Nparticles)
def diffusion(Nsteps):
    for i in range (Nsteps):
        p_index=random.randint(Nparticles)
        direction=random.randint(4)
        if direction==0 and X[p_index]>= - Ngrid+1:
            X[p_index]-=1
        if direction==1 and X[p_index]<=  Ngrid-1:
            X[p_index]+=1
        if direction==2 and Y[p_index]>= - Ngrid+1:
            Y[p_index]-=1
        if direction==3 and Y[p_index]<=  Ngrid-1:
            Y[p_index]+=1
    return [X,Y]  
[X1,Y1]=diffusion(100000)
plt.plot(X1,Y1,'.',color='red',label='small stepsize')
plt.title('Distribution of Particle with time')
plt.ylabel('distribution of particle')
plt.xlabel(' stepsize')
plt.legend()
plt.figure(1)
[X2,Y2]=diffusion(300000)
plt.plot(X2,Y2,'.',color='blue',label='medium stepsize')
plt.legend()
plt.figure(1)
[X3,Y3]=diffusion(500000)
plt.plot(X3,Y3,'.',color='green',label='large stepsize')
plt.legend();

        
    
```

```python
#Problem2: Hole in the Box
import math
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
%matplotlib inline
Ngrid= 100
Nparticles=400
Nsteps=10000
holetop=20
holebot=-20
X= np.zeros(Nparticles)
Y= np.zeros(Nparticles)
particleleft=[]
countparticle=[Nparticles]
for j in range (Nsteps):
    for j in range(Nparticles):
        if j in particleleft:
            continue
        p_index=random.randint(Nparticles)
        direction=random.randint(4)
        if direction==0 and X[p_index]== - Ngrid and holetop > Y[p_index] and holebot < Y[p_index] :
            Nparticles-=1
            particleleft.append(j)
            X[p_index]=None
            Y[p_index]=None
        else:    
            if direction==0 and X[p_index]>= - Ngrid+1:
                X[p_index]-=1
            if direction==1 and X[p_index]<=  Ngrid-1:
                X[p_index]+=1
            if direction==2 and Y[p_index]>= - Ngrid+1:
                Y[p_index]-=1
            if direction==3 and Y[p_index]<=  Ngrid-1:  
                Y[p_index]+=1
        
    countparticle.append(Nparticles)   
plt.plot(countparticle)
plt.ylabel('Number of particle remaining')
plt.xlabel('stepsize')
plt.title('Hole Problem, Number of particle Vs timestep');
```

```python
#Problem3: Hole in the Box for many trials
import math
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
%matplotlib inline
Ngrid= 10
Nparticles=200
Nsteps=100
Nexperiments=30
holetop=5
holebot= -5
Averageparticle=[]
X= np.zeros(Nparticles)
Y= np.zeros(Nparticles)
for steps in range(Nsteps):
    particleleft=[]
    countparticle=[Nparticles]
    for i in range(Nexperiments):
        for j in range(steps):
            for j in range(Nparticles):
                if j in particleleft:
                    continue
            p_index=random.randint(Nparticles)
            direction=random.randint(4)
            if direction==0 and X[p_index]== - Ngrid and holetop > Y[p_index] and holebot < Y[p_index] :
                Nparticles-=1
                particleleft.append(j)
                X[p_index]=None
                Y[p_index]=None
            else:    
                if direction==0 and X[p_index]>= - Ngrid+1:
                    X[p_index]-=1
                if direction==1 and X[p_index]<=  Ngrid-1:
                    X[p_index]+=1
                if direction==2 and Y[p_index]>= - Ngrid+1:
                    Y[p_index]-=1
                if direction==3 and Y[p_index]<=  Ngrid-1:  
                    Y[p_index]+=1
        countparticle.append(Nparticles)
    Averageparticle.append(np.mean(countparticle))
plt.plot(Averageparticle)
plt.ylabel(' Average Number of particle remaining')
plt.xlabel('stepsize')
plt.title('Hole Problem 2, Average Number of particle Vs timestep');

```
